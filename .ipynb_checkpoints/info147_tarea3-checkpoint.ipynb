## Introducción

<img src="data/vecinos.png" width="400">

Los $K$-vecinos es un método clásico y muy sencillo para hacer clasificación de datos, que se basa en predecir la etiqueta de un dato basado en las etiquetas de los datos de entrenamiento que más se le parecen. La siguiente figura describe graficamente los tres pasos del algoritmo

<img src="data/algoritmo.png" width="600">

En este caso es clave definir una noción de distancia entre los ejemplos y también especificar un valor adecuado para $K$, la cantidad de vecinos que influyen en la predicción.

## Formalismo matemático

Sea una base de datos $E = \{(x_j, y_j), j=1, \ldots, N\}$, con $N$ ejemplos donde $x_j \in \mathbb{R}^{D}$ es un atributo d-dimensional e $y_j \in \{0, 1, 2, \ldots, C-1\}$ son sus etiquetas de clase. Sea ahora una segunda base de datos $T = \{(z_i), i=1, \ldots, M\}$ con $M$ ejemplos donde $z_i \in \mathbb{R}^{D}$ es un atributo d-dimensional. Esta base de datos no tiene etiquetas. El objetivo es clasificar los ejemplos de $T$ en base a las etiquetas de los $K$ ejemplos más cercanos de la base de datos $E$


El algoritmo para clasificar el i-esimo elemento de Z es

**Paso 1** Calculamos la distancia entre $z_i$ y cada elemento de $E$ usando

$$
d(z_i, x_j) = \left ( \sum_{d=1}^D  |z_{id} - x_{jd}|^p \right)^{1/p}
$$

que se conoce como [distancia de Minkowski](https://en.wikipedia.org/wiki/Minkowski_distance). Para el caso $p=2$ se recupera la clásica distancia euclidiana.

**Paso 2** Buscamos las $k$ tuplas $(x_k^{(i)}, y_k^{(i)})$ con menor distancia a $z_i$

**Paso 3** Seleccionamos la clase de $z_i$ según

$$
\text{arg}\max_{c=0, 1, \ldots} \sum_{k=1}^K \frac{\mathbb{1}(c=y^{(i)}_k)}{d(z_i, x^{(i)}_k)}
$$

donde 

$$
\mathbb{1}(a=b) = \begin{cases} 1 & \text{si } a=b \\ 0 &  \text{si } a\neq b \end{cases}
$$

se conoce como función indicadora. Esta versión particular del algoritmo se conoce como clasificador de $k$ vecinos ponderado, ya que una menor distancia (mayor cercanía) aumenta el peso del voto

### Parte 2

#### Implementación de una nueva versión de KNN usando Cython

```python
%%cython -l m
import cython
cimport numpy as npc
import numpy as np

# Definicion de tipos de dato
ctypedef npc.float64_t TIPOF_t
ctypedef npc.int64_t TIPOI_t

# Funciones de C
cdef extern from "math.h":
    double pow(double, double)
    double fabs(double)

# Se deshabilitan las comprobaciones de Python    
@cython.boundscheck(False)
@cython.wraparound(False)
@cython.cdivision(True)
def KNN_cython(double [:,::1] X, long [::1] Y, double [:,::1] Z, int k=5, float p=2.):
    cdef long [::1] C = np.unique(Y)
    cdef long N = X.shape[0]
    cdef long D = X.shape[1]
    cdef long M = Z.shape[0]
    dist = np.zeros(shape=(M, N), dtype=np.double)
    cdef double [:,::1] dist_view = dist
    cdef double res
    cdef int i, j, c, n, idx
    for i in range(M):
        for j in range(N):
            dist_view[i, j] = pow((pow(fabs(Z[i,0] - X[j,0]), p) + pow(fabs(Z[i,1] - X[j,1]), p)), 1./p)
    neighbours = np.argsort(dist, axis=1)
    cdef long [:,::1] neighbours_view = neighbours
    neighbours_view = neighbours_view[:,:k]
    Z_Y = np.zeros(shape=(M, ), dtype=np.double)
    cdef double [::1] Z_Y_view = Z_Y
    criterion = np.zeros(shape=(len(C),), dtype=np.double)
    cdef double [::1] criterion_view = criterion
    for i in range(M):
        for c in C:
            idx = 0
            res = 0.0
            for n in neighbours_view[i]:
                if(Y[n] == c):
                    res += 1./dist[i, neighbours_view[i]][idx]
                idx+=1    
            criterion_view[c] = res
        Z_Y_view[i] = np.argmax(criterion)
    return Z_Y
```

#### Verificación de que se obtiene el mismo resultado que con la función original

```python
# Comparación de resultados de la función modificada con la original
print(np.allclose(KNN_cython(x,y,z), KNN(x,y,z)))
```

Luego de implementar la nueva función con Cython usando tipado fijo, vista de arreglos y funciones estádar de la librería matemática de C se logró obtener los mismos resultados que la función KNN original.

#### Grafica de speed-up de la nueva función en comparación a la original

```python
N = [10,50,100,500,1000,5000,10000]
t_original = np.zeros(7)
t_cython = np.zeros(7)
for i in range(7):
    e, t = create_data(N[i])
    x, y = e
    z, w = t
    # Se guardan los objetos TimeitResult de la iteración
    a = %timeit -r3 -n1 -o KNN(x,y,z)
    b = %timeit -r3 -n1 -o KNN_cython(x,y,z)
    # Se guarda el promedio de los objetos en sus respectivos arreglos
    t_original[i] = a.average
    t_cython[i] =  b.average
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,4))
ax1.plot(N, t_original, label="KNN original")
ax1.plot(N, t_cython, label="KNN_cython")
ax1.set_ylabel("Tiempo [s]")
ax1.set_xlabel("Número de ejemplos")
ax1.set_title("Desempeño de los algoritmos")
ax1.legend()
ax2.plot(N, t_original, label="KNN original")
ax2.plot(N, t_cython, label="KNN_cython")
ax2.set_ylabel("Tiempo")
ax2.set_xlabel("Número de ejemplos")
ax2.set_title("Desempeño - Escala logarítmica")
ax2.set_yscale('log')
ax2.legend()
```

Luego de realizar mediciones de tiempo con diferentes tamaños de N, se observa un speed-up muy notorio al utilizar el nuevo algortimo, de hecho, en cada una de las mediciones, el tiempo total de KNN_cython es varias ordenes de magnitud menor al algortimo KNN original. Esta amplia diferencia es tal que opaca totalmente la curva del nuevo algoritmo en el gráfico de la izquierda, por lo que a modo de complemento, se incluye el gráfico con escala logarítmica. Es remarcable notar cómo mediante el uso de funciones matemáticas de C y su eficiencia se pudo reolver el cuello de botella mencionado anteriormente.

#### Validación cruzada para encontrar los mejores valores de  k y p

```python
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
```

```python
kf = KFold(n_splits = 10)
```

Buscando el mejor valor para k

```python
# Valores posible de K
for ki in range(1,20): 
    score = np.zeros(shape=(kf.get_n_splits(), ))
    for i, (train_index, valid_index) in enumerate(kf.split(x)):
        z_y = KNN_cython(x[train_index], y[train_index], z[valid_index], k=ki, p=2.)
        score[i] = accuracy_score(w[valid_index], z_y)
    print("k = ",ki,"   score: ", np.mean(score))
```

Mejor valor k=18

Buscando el mejor valor de p

```python
# Rango de valores de p
p_v = np.linspace(0,4,30)

for pi in p_v: 
    score = np.zeros(shape=(kf.get_n_splits(), ))
    for i, (train_index, valid_index) in enumerate(kf.split(x)):
        z_y = KNN_cython(x[train_index], y[train_index], z[valid_index], k = 18, p=pi)
        score[i] = accuracy_score(w[valid_index], z_y)
    print("p = ",round(pi,4),"   score: ", np.mean(score))
```

El mejor valor para p=1.3793
